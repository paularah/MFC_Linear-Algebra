<p style='text-align:center'><strong>Problem Importance and Background</strong></p>
<p>Balancing of chemical equation is one first few topics students are taught in chemistry. Students like wenseslaus are often taught taught to rely on intelligent guesses and heuristics to balance chemical equations. This method is called the inspection method. Inspection method works for small and relatively easy chemical equations but as they get complex, relying on guesswork to derive the right coefficients to balance a chemical equation can be inefficient and pain staking. As a computer science student, Wenseslaus is on the search for a more algorithimic way to balance chemical equations considering he has to balance lots of complex chemical equations on a daily basis. The balancing of chemical equations can be made much easier, especially for those who find it difficult, by moving the procedures toward the algorithmic and away from the heuristic. That is, a “step to step” procedure is simpler to master than is the haphazard hopping of inspection, even a highly refined inspection (Hutchings et al.,
2007).
</p>

<p>
    A chemical equation is an expression representating of a chemical reaction. The reactants are placed on the left hand side and the products on the right hand side. An arrow indicating if a reaction is reversible or irriversible seperates on the reactants from the products. In the case of a reversible reaction there's one arrow pointing to the right and in the case of an irriversible reaction there is another arrow pointing in the opposite direction indicating that the chemical reaction can go both ways. Each substance taking part in the chemical reaction represented by its molecular formula. A chemical equation is said to be balanced if the number of atoms in each component of the reactants and products are the same.   
</p>
<center>
    $$
    O_2 + H_2 \rightarrow  H_2O
    $$
</center>
<p>
    This is a simple chemical reaction that represents the formation of water from its individual components. On counting the number atoms of the individual elements in this reaction, we realise the equation, is not balanced. To balance this equation using inspection, we add the right coefficients to the ensure the total number of atoms on the reactant side and the product side are the same. 
</p>
<center>
    $$
     O_2 + 2H_2 \rightarrow  2H_2O
    $$
</center>

<p>
    On summing up the number of atoms in this reaction i.e the coefficient multiplied by the subscript for each element, we see that this reaction is now balanced. One mole of oxygen reacts with 2 molesof hydrogen to form water. 
</p>

<p>
    Chemical reaction equations are mathematical and symbolic models of real world components. Therefore it is
governed by specific rules and principles. Law of the Conservation of Matter is one of them, which states that the total
mass of the products must equal the total mass of the reactants. Because of this law, chemical reactions can be thought
of as a system of linear equations and as such can be mathematically manipulated(Abdelrahim, 2017). Now lets gun for a more complex chemical reaction. This is a neutrilisation reaction between an acid and a base to produce a salt and water. At its cuurent state, it is unbalanced. 
</p>
<center>
    $$
    H_2SO_4 + NaOH \rightarrow  Na_2SO_4 + H_2O
    $$
</center>
<p>
    To balance this equation, we assign the unknown coeffients to each substance in this reaction.
</p>
<center>
    $$
    (x_1)NaOH + (x_2)H_2SO_4 \rightarrow (x_3)Na_2SO_4 + (x_4)H_2O
    $$
</center>

<p>
    When we compare the number of atoms for each of components of the equation, we obtain four linear equations and four unknowns. 
</p>
<center>
    $$
    Sodium:    x_1 + 0x_2 = 2x_3 + 0x_4 \\
    Oxygen: x_1 + 4x_2 = 4x_3 + x_4 \\
    Hydrogen: x_1 + 2x_2 = 0x_3 + 2x_4 \\
    Sulphur: 0x_1 + x_2 = x_3 + 0x_4 
    $$
</center>
<p>We can then treat like we would treat any other system of linear equations</p>

<center>
    $$
    x_1 + 0x_2 - 2x_3 - 0x_4 = 0 \\
    x_1 + 4x_2 - 4x_3 - x_4  = 0 \\
    x_1 + 2x_2 - 0x_3 - 2x_4 = 0 \\
    0x_1 + x_2 - x_3 - 0x_4 = 0
    $$
</center>
<p>We then represent this system of linear equations as a matrix</p>
<center>$$\begin{bmatrix} 1 & 0 & -2 & 0   \\ 1 & 4 & -4 & -1  \\ 1 & 2 & 0 & -2 \\ 0 & 1 & -1 & 0 \end{bmatrix}\begin{bmatrix} x_1 \\ x_2 \\ x_3 \\ x_4 \end{bmatrix} =\begin{bmatrix} 0 \\ 0 \\ 0 \\ 0 \end{bmatrix} $$ </center>

<p>We then form the augumented matrix</p>
<center>
    $$
    \left[
      \begin{array}{rrrr|r}
        1 & 0 & -2 & 0 & 0 \\
        1 & 4 & -4 & -1 & 0 \\
        1 & 2 & 0 & -2 & 0 \\
        0 & 1 & -1 & 0 & 0 
      \end{array}
    \right]
    $$
</center>
<p>At this point, we can either use Gaussian elimination to obtain and upper traingular matrix i.e Row echelon form and then  back substitute for each of the unknowns or we perform elementary row operations and tranform our matrix into reduced row echelon form. We achieve this by progressively taking taking a pivot along the diagonal and makking its value 1 and every other value underneath and above 0 using the fundamental row operations</p>
<center>
    $$
    R_2 - R_1 \implies R_2 \left[
      \begin{array}{rrrr|r}
        1 & 0 & -2 & 0 & 0 \\
        0 & 4 & -2 & -1 & 0 \\
        1 & 2 & 0 & -2 & 0 \\
        0 & 1 & -1 & 0 & 0 
      \end{array}
    \right]
    $$
</center>
<p style='text-align:center'v>    
    $$
    R_3 - R_1 \implies R_3 \left[
      \begin{array}{rrrr|r}
        1 & 0 & -2 & 0 & 0 \\
        0 & 4 & -2 & -1 & 0 \\
        0 & 2 & 2 & -2 & 0 \\
        0 & 1 & -1 & 0 & 0 
      \end{array}
    \right]
    $$
</p>
<p>We move to the next pivot along the diagonal and make it 1 and everything else underneath 0 using the row operations</p>

<p style='text-align:center'v>    
    $$
    \frac{1}{4}R_2 - R_2 \implies R_3 \left[
      \begin{array}{rrrr|r}
        1 & 0 & -2 & 0 & 0 \\
        0 & 1 & -\frac{1}{2} & -\frac{1}{4} & 0 \\
        0 & 0 & -3 & \frac{3}{2} & 0 \\
        0 & 1 & -1 & 0 & 0 
      \end{array}
    \right]
    $$
</p>

<p style='text-align:center'>
    $$
    2R_2 - R_3 \implies R_3 \left[
      \begin{array}{rrrr|r}
        1 & 0 & -2 & 0 & 0 \\
        0 & 1 & -\frac{1}{2} & -\frac{1}{4} & 0 \\
        0 & 0 & 6 & -3 & 0 \\
        0 & 1 & -1 & 0 & 0 
      \end{array}
    \right]
    $$
</p>

<p style='text-align:center'> 
    $$
    4R_4 - R_2 \implies R_4 \left[
      \begin{array}{rrrr|r}
        1 & 0 & -2 & 0 & 0 \\
        0 & 4 & -2 & -1 & 0 \\
        0 & 0 & 6 & -3 & 0 \\
        0 & 0 & -2 & 1 & 0 
      \end{array}
    \right]
    $$ 
</p>
